```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
import spacy

data = {
    'Symptoms': ["itchy rash, redness", "scaly patches, inflammation", "raised bumps, itching", "blistering, oozing, pain",
                 "red spots, fever, fatigue", "dark, irregular moles", "dry, itchy skin, thickened plaques",
                 "pink or red rash, circular shape", "small, fluid-filled blisters", "swollen, painful joints, skin rash"],
    'SkinDisease': ["Dermatitis", "Psoriasis", "Hives", "Herpes", "Measles", "Melanoma", "Eczema", "Ringworm", "Chickenpox", "Lupus"]
}

df = pd.DataFrame(data)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['Symptoms'], df['SkinDisease'], test_size=0.2, random_state=42)

# Build a pipeline with a TF-IDF vectorizer and a Naive Bayes classifier
model = make_pipeline(TfidfVectorizer(), MultinomialNB())

# Train the model
model.fit(X_train, y_train)
symptoms_ls=[i.lower() for i in symptoms_list]
# Extracting nouns and adjectives from user input for prediction
user_symptoms = " ".join(symptoms_ls)

# Make a prediction
predicted_disease = model.predict([user_symptoms])

# Display the predicted disease
print("Predicted Skin Disease:", predicted_disease[0])
```
